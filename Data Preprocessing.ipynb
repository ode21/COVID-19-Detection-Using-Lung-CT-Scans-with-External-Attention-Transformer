```python
import os
import random
import pickle
from sklearn.model_selection import train_test_split
import cv2
import numpy as np
```

```python
covid_path = '/content/drive/My Drive/TA_OmarDhani/COVID'
non_covid_path = '/content/drive/My Drive/TA_OmarDhani/non-COVID'

covid_files = [os.path.join(covid_path, f) for f in os.listdir(covid_path) if os.path.isfile(os.path.join(covid_path, f))]
non_covid_files = [os.path.join(non_covid_path, f) for f in os.listdir(non_covid_path) if os.path.isfile(os.path.join(non_covid_path, f))]
```

```python
# Split data into training and testing sets
covid_train, covid_test = train_test_split(covid_files, test_size=0.3, random_state=42)
non_covid_train, non_covid_test = train_test_split(non_covid_files, test_size=0.3, random_state=42)

# Split training data into training and validation sets
covid_train, covid_val = train_test_split(covid_train, test_size=0.3, random_state=42)
non_covid_train, non_covid_val = train_test_split(non_covid_train, test_size=0.3, random_state=42)

# Combine the splits
X_train = covid_train + non_covid_train
X_val = covid_val + non_covid_val
X_test = covid_test + non_covid_test

# Create labels
y_train = [1] * len(covid_train) + [0] * len(non_covid_train)
y_val = [1] * len(covid_val) + [0] * len(non_covid_val)
y_test = [1] * len(covid_test) + [0] * len(non_covid_test)
```

```python
split_data = {
    'X_train': X_train,
    'X_val': X_val,
    'X_test': X_test,
    'y_train': y_train,
    'y_val': y_val,
    'y_test': y_test
}

save_path = '/content/drive/My Drive/TA_OmarDhani/split_data.pkl'

with open(save_path, 'wb') as f:
    pickle.dump(split_data, f)

print(f"Data splits saved to {save_path}")
```

```python
save_path = '/content/drive/My Drive/TA_OmarDhani/split_data.pkl'

with open(save_path, 'rb') as f:
    split_data = pickle.load(f)

X_train = split_data['X_train']
X_val = split_data['X_val']
X_test = split_data['X_test']
y_train = split_data['y_train']
y_val = split_data['y_val']
y_test = split_data['y_test']
```

```python
def load_images(file_paths):
    images = []
    for file_path in file_paths:
        img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (128, 128))  # Resize images to a fixed size
        images.append(img)
    return np.array(images)

X_train = load_images(X_train)
X_val = load_images(X_val)
X_test = load_images(X_test)

# Reshape and normalize the images
X_train = X_train.reshape(-1, 128, 128, 1) / 255.0
X_val = X_val.reshape(-1, 128, 128, 1) / 255.0
X_test = X_test.reshape(-1, 128, 128, 1) / 255.0
```

```python
base_path = '/content/drive/My Drive/TA_OmarDhani/'

with open(base_path + 'X_train.pickle', 'wb') as f:
    pickle.dump(X_train, f)

with open(base_path + 'X_val.pickle', 'wb') as f:
    pickle.dump(X_val, f)

with open(base_path + 'X_test.pickle', 'wb') as f:
    pickle.dump(X_test, f)

with open(base_path + 'y_train.pickle', 'wb') as f:
    pickle.dump(y_train, f)

with open(base_path + 'y_val.pickle', 'wb') as f:
    pickle.dump(y_val, f)

with open(base_path + 'y_test.pickle', 'wb') as f:
    pickle.dump(y_test, f)

print("Files saved successfully in Google Drive.")
```
